```python
# module 1.1 Measures of Central Tendency
# menghitung mean
import statistics

# deklarasi data tinggi badan
data = [100.5, 99.7, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]

# menjumlahkan semua data yang ada
jumlah = sum(data)
print(f"Jumlah data tersebut adalah: {jumlah}")

# menghitung banyaknya data
total = len(data)
print(f"Total data tersebut adalah: {total}")

mean = jumlah / total
print(f"Mean: {mean}")

# diagram grafik mean
import matplotlib.pyplot as plt

plt.hist(data, bins=50, edgecolor='black', alpha=1, color='yellow')

plt.axvline(x=mean, color='blue', linestyle='dashed', linewidth=2, label=f'mean = {mean:.2f}')

plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title('Mean Data')

plt.legend()
plt.show()


# median data
urutan = sorted(data)
print(f"Urutan data tersebut adalah: {urutan}")

data_tengah = total//2

if total % 2 == 0:
    med1 = urutan[data_tengah]
    med2 = urutan[data_tengah - 1]
    median = (med1 + med2) / 2
else:
    median = urutan[data_tengah]
print(f"Median data tersebut: {median}")

# diagram grafik median
import matplotlib.pyplot as plt

plt.hist(data, bins=50, edgecolor='black', alpha=1, color='yellow')
plt.axvline(x=median, color='blue', linestyle='dashed', linewidth=2, label=f'median = {median:.2f}')

plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title('Median Data')

plt.legend()
plt.show()


# menghitung modus data
from collections import Counter

counter = Counter(data)
mod = dict(counter)
paling_sering_muncul = max(list(counter.values()))
print(f"Paling sering muncul: {paling_sering_muncul}")

modus = []
for a, b, in mod.items():
    print(f"{a} terdapat {b} data")
if b == max(list(counter.values())):
    modus.append(a)
    
print(f"Modus data tersebut: {modus}")

# diagram grafik modus
import matplotlib.pyplot as plt

plt.hist(data, bins=50, edgecolor='black', alpha=1, color='yellow')

plt.axvline(x=modus[0], color='blue', linestyle='dashed', linewidth=2, label=f'Modus 1 = {modus[0]:.2f}')

plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title('Modus Data')

plt.legend()
plt.show()
```
