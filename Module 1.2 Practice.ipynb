```python
# modul 1.2 Measures of Variability
# Gallant Damas Hayuaji (5027231037)

# soal 1 Data gol pemain sepak bola
# menghitung range
print("\033[34m" + f"Soal 1: Data Gol Pemain Sepak Bola\n" + "\033[0m")
def hitung_range(data):
    gol_banyak = max(data)
    print(f"Gol paling banyak adalah: {gol_banyak}")
    gol_sedikit = min(data)
    print(f"Gol paling sedikit adalah: {gol_sedikit}")
    range = max(data) - min(data)
    print(f"Range data tersebut: {gol_banyak} - {gol_sedikit} = {range}")
    return range

# memasukkan data gol pemain
data = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
print(f"Jumlah gol: {data}")

range = hitung_range(data)

# display range gol dalam grafik
import matplotlib.pyplot as plt

plt.hist(data, bins=50, edgecolor='black', alpha=1, color='red')

y = 1
x1 = 4/70*1
x2 = 30/70*1

plt.axhline(y=y, xmin=x1, xmax=x2, color='green', linestyle='dashed', linewidth=1, label=f'Range = {hitung_range(data)}')

plt.axvline(x=min(data), color='blue', linestyle='dashed', linewidth=1, label=f'Min Data: {min(data):.2f}')
plt.axvline(x=min(data), color='yellow', linestyle='dashed', linewidth=1, label=f'Max Data: {max(data):.2f}')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title('Range')
                
plt.legend()
plt.show()

def rata_rata(data):
    return sum(data) / len(data)

def hitung_variansi(data):
    mean = rata_rata(data)
    print(f"Rata-rata data tersebut: {mean}")
    
# mencari kuadrat selisih tiap data gol dengan rata-rata
    print(f"Kuadrat selisih data gol dengan rata-rata:")
    squared_diff = []
    for n in data:
        squared_diff_per_n = (n - mean)**2
        squared_diff.append(squared_diff_per_n)
        print(f"({n} - {mean})**2 = {squared_diff_per_n}")
        
        
        total_squared_diff = sum(squared_diff)
    print(f"Total kuadrat selisih data: {total_squared_diff}")
    print(f"Banyaknya data gol: {len(data)}")
    
    # Menghitung variansi
    variansi = total_squared_diff / len(data)
    print(f"Variansi data tersebut: {total_squared_diff} / {len(data)} = {variansi}")
    
    return variansi

# Memasukkan data gol pemain untuk dihitung variansinya
data = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
print(f"Jumlah gol: {data}")
variansi = hitung_variansi(data)
print(f"\n")

# menghitung standar deviasi
def hitung_sd(data):
    variansi = hitung_variansi(data)
    sd = variansi**0.5
    print(f"Standar deviasi data tersebut: {variansi}**0.5 = {sd}")
    return sd

# Memasukkan data gol pemain untuk dihitung standar deviasinya
data = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
print(f"Jumlah gol: {data}")

sd = hitung_sd(data)
mean = rata_rata(data)

# display standar deviasi dalam grafik
import matplotlib.pyplot as plt
import numpy as np

plt.hist(data, bins=50, edgecolor='black', alpha=0.5, color='green')

plt.axvline(x=mean, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean}')
plt.axvline(x=mean - sd, color='blue', linestyle='dashed', linewidth=1, label=f'{mean-sd:.2f}')
plt.axvline(x=mean + sd, color='yellow', linestyle='dashed', linewidth=1, label=f'{mean+sd:.2f}')

plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title(f'Standar Deviasi = {sd:.2f}')

plt.legend()
plt.show()
print("Ukuran sebaran data yang paling mampu menggambarkan sebaran data terbaik adalah standar deviasi yaitu sebesar 3.03. Standar deviasi lebih sensitif terhadap perubahan data dan menampilkan satuan yang sama dengan data asli.")
print(f"\n")

# Soal 2 Data Tinggi Badan
# Menghitung sebaran data range, variansi, dan standar deviasi
print("\033[34m" + f"Soal 2: Tinggi Badan Tim Olahraga\n" + "\033[0m")

# menghitung range tinggi badan
def range_tb(data):
    tb_tinggi = max(data)
    print(f"Pemain paling tinggi adalah: {tb_tinggi}")
    tb_pendek = min(data)
    print(f"Pemain paling pendek adalah: {tb_pendek}")
    range = max(data) - min(data)
    print(f"Range tinggi badan pemain adalah: {tb_tinggi} - {tb_pendek} = {range}")
    return range

# memasukkan data tinggi badan pemain
data = [180, 185, 175, 190, 170, 195, 165]
print(f"Total tinggi badan: {data}")
range = range_tb(data)

# display data tinggi badan dalam grafik
import matplotlib.pyplot as plt

plt.hist(data, bins=100, edgecolor='black', alpha=1, color='red')

y = 1
x1 = 4/70*1
x2 = 30/70*1

plt.axhline(y=y, xmin=x1, xmax=x2, color='green', linestyle='dashed', linewidth=1, label=f'Range = {range_tb(data)}')
plt.axvline(x=min(data), color='blue', linestyle='dashed', linewidth=1, label=f'Min Data: {min(data):.2f}')
plt.axvline(x=min(data), color='yellow', linestyle='dashed', linewidth=1, label=f'Max Data: {max(data):.2f}')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title('Range')

plt.legend()
plt.show()

def rata_tb(data):
    return sum(data) / len(data)

def variansi_tb(data):
    mean = rata_tb(data)
    print(f"Rata-rata tinggi badan pemain adalah: {mean}")
    
# menghitung kuadrat selisih tiap data gol dengan rata-rata
    print(f"Kuadrat selisih tinggi badan dengan rata-rata:")
    kuadrat_selisih = []
    for n in data:
        kuadrat_selisih_per_n = (n - mean)**2
        kuadrat_selisih.append(kuadrat_selisih_per_n)
        print(f"({n} - {mean})**2 = {kuadrat_selisih_per_n}")
        
        total_kuadrat_selisih = sum(kuadrat_selisih)
    print(f"Total kuadrat selisih tinggi badan: {total_kuadrat_selisih}")
    print(f"Banyaknya data tinggi badan: {len(data)}")
    
# menghitung variansi
    variansi = total_kuadrat_selisih / len(data)
    print(f"Variansi tinggi badan adalah: {total_kuadrat_selisih} / {len(data)} = {variansi}")
    
    return variansi
# memasukkan data tinggi badan untuk dihitung variansinya
data = [180, 185, 175, 190, 170, 195, 165]
print(f"Total tinggi badan: {data}")
variansi = variansi_tb(data)
print(f"\n")

# menghitung standar deviasi tinggi badan
def sd_tb(data):
    variansi = variansi_tb(data)
    deviasi_tb = variansi**0.5
    print(f"Standar deviasi tinggi badan tersebut: {variansi}**0.5 = {deviasi_tb}")
    return deviasi_tb

# memasukkan data tinggi badan untuk dihitung standar deviasi
data = [180, 185, 175, 190, 170, 195, 165]
print(f"Jumlah tinggi badan: {data}")

deviasi_tb = sd_tb(data)
mean = rata_tb(data)

# menampilkan standar deviasi dalam grafik
import matplotlib.pyplot as plt
import numpy as np

plt.hist(data, bins=100, edgecolor='black', alpha=0.5, color='green')

plt.axvline(x=mean, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean}')
plt.axvline(x=mean - sd, color='blue', linestyle='dashed', linewidth=1, label=f'{mean-deviasi_tb:.2f}')
plt.axvline(x=mean + sd, color='yellow', linestyle='dashed', linewidth=1, label=f'{mean+deviasi_tb:.2f}')

plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title(f'Standar Deviasi = {deviasi_tb:.2f}')

plt.legend()
plt.show()
print("Ukuran sebaran data yang paling mampu menggambarkan data terbaik adalah standar deviasi yaitu sebesar 10.00. Standar deviasi memiliki satuan yang sama dengan data asli dan standar deviasi lebih sensitif terhadap perubahan data.")
```
